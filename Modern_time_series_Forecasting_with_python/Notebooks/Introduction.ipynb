- Time series is a set of observations taken sequentially in time.
- There are two types of time series data based on time intervals.
    - **`Regular time series`** - this is the most commont type of time series, where we have observations coming in at regular intervals of time.
    - **`Irregular time series`** - There are few time series where we do not have observations at regular intervals of time.
- Main areas of application for time series analysis
    - **`Time series forecasting`** - Predicting the future values of a time series, given the past values.
    - **`Time series calssification`** - Sometimes instead of predicting the future value fo the time series, we may also want to predict an action based on past values.
    - **`Outlier Detection`** - There are some situations where we only want to detect if something is going wrong or if something is out of the ordinary.
    - **`Interpretation and causality`** - You can use time-series analysis to understand the whats and whys of the time series based on past values, understand the relationships between several related time series, or derive causal inferences based on time series data.

### **`Data-Generating process(DGP)`** 
- In statistics, this underlying process that generates the time series is referred to as to DGP. Time series data is produced by stochastic and deterministic process.
- A stochastic Process is way to describe how things change over time in a random but somewhat predictanble manner.
